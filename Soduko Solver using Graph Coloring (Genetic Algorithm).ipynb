```python
from array import *
import random
import matplotlib.pyplot as plt
import numpy as np
Gen = np.array([])
Fit = np.array([])

sudoku = [[7,8,0, 4,0,0, 1,2,0],
		[6,0,0, 0,7,5, 0,0,9],
		[0,0,0, 6,0,1, 0,7,8],
		[0,0,7, 0,4,0, 2,6,0],
		[0,0,1, 0,5,0, 9,3,0],
		[9,0,4, 0,6,0, 0,0,5],
		[0,7,0, 3,0,0, 0,1,2],
		[1,2,0, 0,0,7, 4,0,0],
		[0,4,9, 2,0,6, 0,0,7]]

def pretty_print(grid):
    side = len(grid)
    bottom = int(side ** 0.5)
    expand_line = lambda line : line[0]+line[5:9].join([line[1:5]*(bottom-1)]*bottom)+line[9:13]
    # line zero 
    l_0 = expand_line("╔═══╤═══╦═══╗")
    # line one 
    l_1 = expand_line("║ . │ . ║ . ║")
    # line two
    l_2 = expand_line("╟───┼───╫───╢")
    # line three
    l_3 = expand_line("╠═══╪═══╬═══╣")
   # line four  
    l_4 = expand_line("╚═══╧═══╩═══╝")
    symb = " 1234567890ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    numbers = [[""] + [symb[n] for n in row] for row in grid]
    print(l_0)
    for r in range(1, side + 1):
        print("".join(n + s for n, s in zip(numbers[r - 1], l_1.split("."))))
        print([l_2, l_3, l_4][(r % side == 0) + (r % bottom == 0)])
        
        
pretty_print(sudoku)

def make_sudoku(candidate):
    matrix = [[0 for _ in range(9)] for _ in range(9)]
    for n in range(len(candidate)):
        x, y = convert(n)
        matrix[x][y] = candidate[n]
    
    return matrix
  
'''Create Graph'''

# Define the size of the Sudoku grid
size = 9

# Initialize the adjacency matrix with all zeros
adj_matrix = [[0 for _ in range(size**2)] for _ in range(size**2)]

# Define a function to get the index of a cell in the adjacency matrix
def get_index(row, col):
    return row*size + col

# Populate the adjacency matrix
for row in range(size):
    for col in range(size):
        cell_index = get_index(row, col)
        # Add edges to cells in the same row
        for i in range(size):
            adj_matrix[cell_index][get_index(row, i)] = 1
        # Add edges to cells in the same column
        for j in range(size):
            adj_matrix[cell_index][get_index(j, col)] = 1
        # Add edges to cells in the same box
        box_row_start = (row // 3) * 3
        box_col_start = (col // 3) * 3
        for i in range(box_row_start, box_row_start + 3):
            for j in range(box_col_start, box_col_start + 3):
                adj_matrix[cell_index][get_index(i, j)] = 1
n = 81

for i in range(n):
    for j in range(n):
        if i == j:
            adj_matrix[i][j] = 0

graph = adj_matrix
print(graph)

def convert(n):
    i = int(n/9)
    j = n%9
    return i,j


# graph = Adj
'''Upper Bound for Coloring'''
max_num_colors = 9
  
  
'''Create Individual using given # of colors'''
number_of_colors = max_num_colors
'''GA'''
condition = True
while(condition and number_of_colors > 0):
    def create_individual():
        individual = []
        for i in range(n):
            x ,y = convert(i)
            if sudoku[x][y] != 0:
                individual.append(sudoku[x][y])
            else:
                individual.append(random.randint(1, number_of_colors))
        return individual
    '''Create Population'''
    population_size = 1000
    generation = 0
    population = []
    for i in range(population_size):
        individual = create_individual()
        population.append(individual)
  
    '''Fitness'''
    def fitness(graph, individual):
        fitness = 0
        for i in range(n):
            for j in range(i, n):
                if(individual[i] == individual[j] and graph[i][j] == 1):
                    fitness += 1
        return fitness
  
    '''Crossover'''
    def crossover(parent1, parent2):
        position = random.randint(2, n-2)
        child1 = []
        child2 = []
        for i in range(position+1):
            child1.append(parent1[i])
            child2.append(parent2[i])
        for i in range(position+1, n):
            child1.append(parent2[i])
            child2.append(parent1[i])
        return child1, child2
  
    def mutation1(individual):
        probability = 0.6
        check = random.uniform(0, 1)
        if(check <= probability):
            position = random.randint(0, n-1)
            x, y = convert(position)
            if sudoku[x][y] != 0:
                return individual
            individual[position] = random.randint(1, number_of_colors)
        return individual
  
    def mutation2(individual):
        probability = 0.2
        check = random.uniform(0, 1)
        if(check <= probability):
            adjacentColors = []
            position = random.randint(0, n - 1)
            for i in range (n):
                if graph[position][i] == 1:
                    adjacentColors.append(individual[i])
            if individual[position] not in adjacentColors:
                return individual
            candidateColors = list(set(list(range(1, number_of_colors + 1))) - set(adjacentColors))
            if len(candidateColors) != 0:
                individual[position] = random.choice(candidateColors)
        return individual;
         
    def parent_selection1(population):
        new_population = []
        for j in range(2):
            random.shuffle(population)
            for i in range(0, population_size-1, 2):
                if fitness(graph, population[i]) < fitness(graph, population[i+1]):
                    new_population.append(population[i])
                else:
                    new_population.append(population[i+1])
        return new_population
    
    def parent_selection2(population):
        return population[0], population[1]
        
      
    
    best_fitness = fitness(graph, population[0])
    fittest_individual = population[0]
    gen = 0
    while(best_fitness != 0 and gen != 10000):
        gen += 1
        if random.randint(1,2) ==1:
            parents = parent_selection1(population)
        else :
            parents = parent_selection2(population)
        new_population = population
        
        for i in range(0, len(parents) - 1, 2):
            child1, child2 = crossover(parents[i], parents[i+1])
            new_population.append(child1)
            new_population.append(child2)
            
        for individual in new_population:
            if random.randint(1,2) ==1:
                individual = mutation1(individual)
            else:
                individual = mutation2(individual)
        
        population = new_population
        population = sorted(population, key = lambda x: fitness(graph, x))
        population = population[0: population_size]
        
        best_fitness = fitness(graph, population[0])
        fittest_individual = population[0]
        for individual in population:
            if(fitness(graph, individual) < best_fitness):
                best_fitness = fitness(graph, individual)
                fittest_individual = individual
        if gen % 10 == 0:
            print("Generation: ", gen, "Best_Fitness: ",
                  best_fitness, "Individual: ", fittest_individual)
        if gen % 100 == 0:
            pretty_print(make_sudoku(fittest_individual))
        Gen = np.append(Gen, gen)
        Fit = np.append(Fit, best_fitness)
        
    print("Using ", number_of_colors, " colors : ")
    print("Generation: ", gen, "Best_Fitness: ",
          best_fitness, "Individual: ", fittest_individual)
    print("\n\n")
    
    if(best_fitness != 0):
        condition = False
        print("Graph is ", number_of_colors+1, " colorable")
    else:
        Gen = np.append(Gen, gen)
        Fit = np.append(Fit, best_fitness)
        plt.plot(Gen, Fit)
        plt.xlabel("generation")
        plt.ylabel("best-fitness")
        plt.show()
        Gen = []
        Fit = []
        pretty_print(make_sudoku(fittest_individual))
        break;
```
